```python
import numpy as np
from keras.datasets import mnist
from keras.layers import Dense,Flatten,Conv2D
from keras.models import Sequential
from keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
(X_train,Y_train),(X_test,Y_test)=mnist.load_data()
```

```python
#Y_test
X_train.shape
```

```python
#checking out the shapes involved in dataset
print("Shape of X_train: {}".format(X_train.shape))
print("Shape of Y_train: {}".format(Y_train.shape))
print("Shape of X_test: {}".format(X_test.shape))
print("Shape of Y_test: {}".format(Y_test.shape))
```

```python
X_train=X_train.reshape(60000,28,28,1)
X_test=X_test.reshape(10000,28,28,1)
```

```python
print("Shape of X_train: {}".format(X_train.shape))
print("Shape of Y_train: {}".format(Y_train.shape))
print("Shape of X_test: {}".format(X_test.shape))
print("Shape of Y_test: {}".format(Y_test.shape))
```

```python
Y_train[0:10]
```

```python
X_train[2]
```

```python
Y_train=to_categorical(Y_train)
Y_test=to_categorical(Y_test)
```

```python
#Y_test
Y_train[0:10]
```

```python
X_train=X_train.astype('float32')
X_test=X_test.astype('float32')
```

```python
X_train[0]
```

```python
X_train/=255
X_test/=255
```

```python
X_train[0]
```

```python
import math
numbers_to_display=45
num_cells=math.ceil(math.sqrt(numbers_to_display))
plt.figure(figsize=(20,20))
for i in range(numbers_to_display):
    plt.subplot(num_cells,num_cells,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_train[i],cmap=plt.cm.binary)
    plt.xlabel(Y_train[i])
plt.show()
```

```python
model=Sequential()
layer_1=Conv2D(32,kernel_size=3,activation='relu',input_shape=(28,28,1))
layer_2=Conv2D(64,kernel_size=3,activation='relu')
layer_3=Flatten()
layer_4=Dense(10,activation='softmax')
model.add(layer_1)
model.add(layer_2)
model.add(layer_3)
model.add(layer_4)
```

```python
model.summary()
```

```python
from sklearn.model_selection import train_test_split
X_trn,X_val,Y_trn,Y_val=train_test_split(X_train,Y_train,test_size=0.2,random_state=0)
```

```python
X_trn.shape
```

```python
X_val.shape
```

```python
(_, IMAGE_WIDTH, IMAGE_HEIGHT, IMAGE_CHANNELS)=X_train.shape
IMAGE_CHANNELS=1

print('IMAGE_WIDTH:',IMAGE_WIDTH);
print('IMAGE_HEIGHT:',IMAGE_HEIGHT);
print('IMAGE_CHANNELS:',IMAGE_CHANNELS);
x_train_with_channels=X_trn.reshape(
    X_trn.shape[0],
    IMAGE_WIDTH,
    IMAGE_HEIGHT,
    IMAGE_CHANNELS
)
x_validation_with_channels=X_val.reshape(
    X_val.shape[0],
    IMAGE_WIDTH,
    IMAGE_HEIGHT,
    IMAGE_CHANNELS
)
x_test_with_channels=X_test.reshape(
    X_test.shape[0],
    IMAGE_WIDTH,
    IMAGE_HEIGHT,
    IMAGE_CHANNELS
)
print('x_train_with_channels:',x_train_with_channels.shape)
print('x_validation_with_channels:',x_validation_with_channels.shape)
print('x_test_with_channels:',x_test_with_channels.shape)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
result=model.fit(X_trn,Y_trn,validation_data=(X_val,Y_val),epochs=2)
```

```python
plt.xlabel('Epoch Number')
plt.ylabel('Loss')
plt.plot(result.history['loss'],label='training set')
plt.plot(result.history['val_loss'],label='validation set')
plt.legend()
```

```python
plt.xlabel('Epoch Number')
plt.ylabel('Accuracy')
plt.plot(result.history['accuracy'],label='training set')
plt.plot(result.history['val_accuracy'],label='validation set')
plt.legend()
```

```python
model.evaluate(X_test,Y_test)
```

```python
Y_test=np.argmax(Y_test,axis=1)
Y_pred=model.predict(X_test)
Y_pred=np.argmax(Y_pred,axis=1)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

cm=confusion_matrix(Y_test,Y_pred)
print('Accuracy Score:',accuracy_score(Y_test,Y_pred))
print('Confusion matrix \n',confusion_matrix(Y_test,Y_pred))
print('Classification \n',classification_report(Y_test,Y_pred))
```
