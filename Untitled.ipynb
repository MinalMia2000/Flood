```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calculate_vif(dataset):
    vif=pd.DataFrame()
    vif["features"]=dataset.columns
    vif["VIF_Value"] = [variance_inflation_factor(dataset.values, i) for i in range(dataset.shape[1])]
    return(vif)
```

```python
features=x.iloc[:,:-1]
print(features.head())
calculate_vif(features)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(dataset):
    vif = pd.DataFrame()
    vif["features"] = dataset.columns
    vif["VIF_Value"] = [variance_inflation_factor(dataset.values, i) for i in range(dataset.shape[1])]
    return vif
# Example usage
features = x.iloc[:,:-1]  # Assuming x is your dataset
vif_result = calculate_vif(features)
# Sort VIF values in ascending order
sorted_vif_result = vif_result.sort_values(by='VIF_Value', ascending=True)
# Plotting bar plot
plt.figure(figsize=(10,6))
plt.barh(sorted_vif_result['features'], sorted_vif_result['VIF_Value'], color='skyblue')
plt.xlabel('VIF Value',fontsize=14)
plt.ylabel('Name of The Features',fontsize=14)
plt.title('Variance Inflation Factor (VIF)',fontsize=20)
plt.show()
```
